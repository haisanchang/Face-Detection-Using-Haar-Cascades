```python
import cv2
import sys
import imutils
from IPython.display import Image
# imagePath = sys.argv[1]
# cascPath = sys.argv[2]
imagePath = r'C:\Users\user\Desktop\test.jpg'
cascPath = r'C:\Users\user\Desktop\haarcascade_frontalface_default.xml'
```

```python
# 初始化Haar cascade
face_cascade = cv2.CascadeClassifier(cascPath)
# 讀入欲辨識人臉圖片
image = cv2.imread(imagePath)
# 轉灰階
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(25, 25),
    flags = cv2.CASCADE_SCALE_IMAGE)

# x,y矩形位置、w,h為寬和高
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
new_pic = imutils.resize(image, height = 640)    
    
cv2.imshow("Display faces", new_pic)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
